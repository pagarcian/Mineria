### Primer punto. Uso de colecciones

Tenemos cuatro tipos de colecciones, tuplas, listas, conjuntos y diccionarios. Para comenzar, sabemos que las tuplas se podrian ver como un subconjunto de las listas, pero a diferencia de ellas, estas (tuplas) son inmutables.

#### a) Se desea conocer el promedio, valor maximo y valor minimo de las calorias de cuatro gaseosas muy conocidas en el mercado internacional.

```python
import statistics
```

```python
Calorias=(150,140,160,140)
Gaseosas=("Pepsi","CocaCola","Fanta","Sprite")
min = max = Calorias[0]
for Caloria in Calorias:
    if Caloria < min:
        min = Caloria
    elif Caloria > max:
        max = Caloria 
mean = statistics.mean(Calorias)        
        

print('Por cada 355ml')
for i in range(len(Calorias)):
    print(Gaseosas[i],"tiene",Calorias[i], "Kcal")
    
print()


print("La menor cantidad de calorias encontrada fue de " + str(min)+"Kcal por porción") 
print("La mayor cantidad de calorias encontrada fue de " + str(max)+"Kcal por porción")


print("El promedio de calorias por cada 355ml de las gaseosas es de ", mean, "calorias")
```

#### b) En un pequeño negocio, la dueña quiere comenzar a darle facturas a sus clientes y usarla como mecanismo para calcular el precio final de la compra. Ya que los precios de los productos de la canasta del hogar cambian cada año, la dueña prefiere poder determinar el precio para cada producto al momento del checkout. 

```python
carrito = {}
pendiente = 0
continuar = True
while continuar:
    item = input('Nombre del artículo: ')
    precio = int(input('Introduce el precio de ' + item + ': '))
    carrito[item] = precio
    continuar = input('¿Quieres añadir artículos a la lista (Si/No)? ') == "Si"
coste = 0
print('Lista de la compra')
for item, precio in carrito.items():
    print(item, '\t', precio)
    coste += precio
    pendiente += coste
    print('Total: ', pendiente)
```
